```python
import keras
```

```python

from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD, Adam, RMSprop,Adadelta
from keras.utils import np_utils
```

```python
batch_size = 128
nb_classes = 10
nb_epoch = 20
```

```python
model = Sequential()
model.add(Dense(1024, input_shape=(330,)))
model.add(Activation('relu'))
#model.add(Dropout(0.2))
model.add(Dense(1024))
model.add(Activation('relu'))
#model.add(Dropout(0.2))
model.add(Dense(1))
model.add(Activation('linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',
              optimizer=Adadelta(),
              metrics=['mean_squared_error'])
```

```python
import pandas as pd
```

```python
data =  pd.DataFrame.from_csv("train.csv",index_col=None)

X = data[data.columns[1:]].values
```

```python
y = data["Protein"].values
```

```python
model.fit(X,y,nb_epoch=100)
```

```python
test_data = pd.DataFrame.from_csv("test.csv",index_col=None)

X_test = test_data[test_data.columns[1:]].values
y_test = test_data["Protein"].values
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,model.predict(X_test))
```
